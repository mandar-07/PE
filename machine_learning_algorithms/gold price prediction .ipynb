```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
gold_data = pd.read_csv(r'D:\clg\New folder\New folder\ gld_price_data')
```

```python
gold_data.head()
```

```python
gold_data.tail()
```

```python
gold_data.isnull().sum()
```

```python
X = gold_data.drop(['Date','GLD'],axis=1)
Y = gold_data['GLD']
```

```python
X.head()
```

```python
Y.head()
```

```python
plt.plot(X.SLV, color='blue', label = 'Actual Value')
plt.plot(Y, color='black', label = 'Actual Value')
plt.plot(X.USO, color='red', label = 'Actual Value')
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
regressor.fit(X_train,Y_train)
```

```python
test_data_prediction = regressor.predict(X_test)
```

```python
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared error : ", error_score)
```

```python
import sklearn.metrics as sm
print("Mean absolute error =", round(sm.mean_absolute_error(Y_test, test_data_prediction), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(Y_test, test_data_prediction), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(Y_test,test_data_prediction ), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(Y_test, test_data_prediction), 2)) 
```

```python
Y_test = list(Y_test)
```

```python
plt.plot(Y_test, color='red', label = 'Actual Value')
plt.plot(test_data_prediction, color='green', label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
#plt.legend()
plt.show()
```
